# This notebook makes a 17g6a3 pPb vs 16c3b pp Monte-carlo comparison

```python
import ROOT
from ROOT import TH1, TCanvas, TFile, TLatex, TLegend, TMath
```

```python
# Load in the file from 17g6a3
MCData = TFile("clusterJet_config_17g6a3_pthat1_17g6a3_pthat2_17g6a3_pthat3_17g6a3_pthat4.root", "READ")

# Load in the file from 16c3b
ppMCData = TFile("clusterJet_config_16c3b_pthat1_16c3b_pthat2_16c3b_pthat3_16c3b_pthat4.root", "READ")
```

```python
#Collect the data from 17g6a3
Xj_MC = MCData.Get("hjet_Xj")
Xjtruth_MC = MCData.Get("hjet_Xj_truth")

dPhi_MC = MCData.Get("hjet_dPhi")
dPhitruth_MC = MCData.Get("hjet_dPhi_truth")

ptD_MC = MCData.Get("hjet_ptD")

Multiplicity_MC = MCData.Get("hjet_Multiplicity")

dEta_MC = MCData.Get("hjet_dEta")
dEtatruth_MC = MCData.Get("hjet_dEta_truth")

AvgEta_MC = MCData.Get("hjet_AvgEta")
AvgEtatruth_MC = MCData.Get("hjet_AvgEta_truth")

leading_pT_MC = MCData.Get("hjet_leading_pT")
leading_pTtruth_MC = MCData.Get("hjet_leading_pT_truth")

leading_Eta_MC = MCData.Get("hjet_leading_Eta")
leading_Etatruth_MC = MCData.Get("hjet_leading_Eta_truth")

leading_Phi_MC = MCData.Get("hjet_leading_Phi")
leading_Phitruth_MC = MCData.Get("hjet_leadingPhi_truth")

# Collect the data from 16c3b
Xj_ppMC = ppMCData.Get("hjet_Xj")
Xjtruth_ppMC = ppMCData.Get("hjet_Xj_truth")

dPhi_ppMC = ppMCData.Get("hjet_dPhi")
dPhitruth_ppMC = ppMCData.Get("hjet_dPhi_truth")

ptD_ppMC = ppMCData.Get("hjet_ptD")

Multiplicity_ppMC = ppMCData.Get("hjet_Multiplicity")

dEta_ppMC = ppMCData.Get("hjet_dEta")
dEtatruth_ppMC = ppMCData.Get("hjet_dEta_truth")

AvgEta_ppMC = ppMCData.Get("hjet_AvgEta")
AvgEtatruth_ppMC = ppMCData.Get("hjet_AvgEta_truth")

leading_pT_ppMC = ppMCData.Get("hjet_leading_pT")
leading_pTtruth_ppMC = ppMCData.Get("hjet_leading_pT_truth")

leading_Eta_ppMC = ppMCData.Get("hjet_leading_Eta")
leading_Etatruth_ppMC = ppMCData.Get("hjet_leading_Eta_truth")

leading_Phi_ppMC = ppMCData.Get("hjet_leading_Phi")
leading_Phitruth_ppMC = ppMCData.Get("hjet_leadingPhi_truth")
```

```python
# Graph Xj
Xj_ppMC.SetTitle("Xj distribution; Xj; 1/N dN/dXj")
Xj_ppMC.SetMaximum(0.02)
Xj_ppMC.GetYaxis().SetTitleOffset(1)
Xj_ppMC.GetYaxis().SetLabelSize(0.02)
Xj_ppMC.GetYaxis().SetTitleSize(0.04)
Xj_ppMC.SetLineColor(1)
Xj_ppMC.SetMarkerColor(1)
Xj_ppMC.SetMarkerStyle(20)
Xj_MC.SetLineColor(2)
Xj_MC.SetMarkerColor(2)
Xj_MC.SetMarkerStyle(20)
Xjtruth_MC.SetLineColor(3)
Xjtruth_MC.SetMarkerColor(3)
Xjtruth_MC.SetMarkerStyle(20)
Xjtruth_ppMC.SetLineColor(4)
Xjtruth_ppMC.SetMarkerColor(4)
Xjtruth_ppMC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1000, 1000)
Xj_ppMC.Draw()
Xj_MC.Draw("same")
Xjtruth_MC.Draw("HIST same")
Xjtruth_ppMC.Draw("HIST same")
c.Draw()

legend = TLegend(0.78, 0.77, 0.98, .94)
legend.AddEntry(Xj_MC, "pPb MC results", "l")
legend.AddEntry(Xjtruth_MC, "pPb MC truth", "l")
legend.AddEntry(Xj_ppMC, "pp MC results", "l")
legend.AddEntry(Xjtruth_ppMC, "pp MC truth", "l")
legend.Draw()
c.SaveAs("Xj_MCcomparison.png")
```

```python
# Graph dEta
dEta_ppMC.SetTitle("#Delta #eta distribution; #Delta #eta; 1/N dN/d#Delta#eta")
dEta_ppMC.SetMaximum(0.4)
dEta_ppMC.GetYaxis().SetTitleOffset(1)
dEta_ppMC.GetYaxis().SetLabelSize(0.02)
dEta_ppMC.GetYaxis().SetTitleSize(0.04)
dEta_ppMC.GetXaxis().SetRangeUser(0, 1.2)
dEta_ppMC.SetLineColor(1)
dEta_ppMC.SetMarkerColor(1)
dEta_ppMC.SetMarkerStyle(20)
dEta_MC.SetLineColor(2)
dEta_MC.SetMarkerColor(2)
dEta_MC.SetMarkerStyle(20)
dEtatruth_MC.SetLineColor(3)
dEtatruth_MC.SetMarkerColor(3)
dEtatruth_MC.SetMarkerStyle(20)
dEtatruth_ppMC.SetLineColor(4)
dEtatruth_ppMC.SetMarkerColor(4)
dEtatruth_ppMC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1000, 1000)
dEta_ppMC.Draw()
dEta_MC.Draw("same")
dEtatruth_MC.Draw("HIST same")
dEtatruth_ppMC.Draw("HIST same")
c.Draw()

legend = TLegend(0.78, 0.77, 0.98, .94)
legend.AddEntry(Xj_MC, "pPb MC results", "l")
legend.AddEntry(Xjtruth_MC, "pPb MC truth", "l")
legend.AddEntry(Xj_ppMC, "pp MC results", "l")
legend.AddEntry(Xjtruth_ppMC, "pp MC truth", "l")
legend.Draw()
c.SaveAs("dEta_MCcomparison.png")
```
