# Gets the 13def cutflow graphs

```python
import ROOT
from ROOT import TH1, TH2, TCanvas, TLatex, TLegend, TH2D, TFile
```

```python
file_jetpT_10_0 = TFile("GammaJet_config_clusptmin15.0_clusptmax20.0_JETPTMIN_10.0_DATANAME__13def_4L.root", "READ")
file_jetpT_7_5 = TFile("GammaJet_config_clusptmin15.0_clusptmax20.0_JETPTMIN_7.5_DATANAME__13def_4L.root", "READ")
file_jetpT_5_0 = TFile("GammaJet_config_clusptmin15.0_clusptmax20.0_JETPTMIN_5.0_DATANAME__13def_4L.root", "READ")

evtcutflow_10_0 = file_jetpT_10_0.Get("EventCutFlow")
evtcutflow_7_5 = file_jetpT_7_5.Get("EventCutFlow")
evtcutflow_5_0 = file_jetpT_5_0.Get("EventCutFlow")

cluscutflow_10_0 = file_jetpT_10_0.Get("ClusterCutFlow")
cluscutflow_7_5 = file_jetpT_7_5.Get("ClusterCutFlow")
cluscutflow_5_0 = file_jetpT_5_0.Get("ClusterCutFlow")
```

```python
cluscutflow_10_0.SetTitle("Event cutflow, jet p_{T} > 10 GeV; ; Counts")
#EMCalRatio1D.SetMaximum(1.2)
#EMCalRatio1D.SetMinimum(0.0)
#cluscutflow_10_0.SetMarkerColor(2)
#cluscutflow_10_0.SetLineColor(2)

c = TCanvas("c", "c", 1600, 1600)
evtcutflow_10_0.Draw()
c.Draw()
c.SaveAs("eventcutflow.png")
```
