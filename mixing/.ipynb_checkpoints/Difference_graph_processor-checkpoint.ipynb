```python
import ROOT
from ROOT import TH1, TCanvas, TFile, TLatex, TLegend, TMath
import matplotlib.pyplot as plt
```

```python
# Get file
pairedTrack = TFile("13d_0GeVTrack_paired_0GeVTracks_Correlation_0_to_300.root", "READ")
```

```python
# Get histograms
zVertexdiff = pairedTrack.Get("z_Vertices;1")
Multdiff = pairedTrack.Get("mult_Vertices;1")

#ROOT.gStyle.SetOptStat("")
```

```python
zVertexdiff.SetTitle("Mixed-event z-Vertex differences; z-Vertex (cm); Count")
#zVertexdiff.GetXaxis().SetRangeUser(0, 5)
zVertexdiff.GetYaxis().SetTitleOffset(1.5)
#zVertexdiff.GetYaxis().SetLabelSize(0.04)
#zVertexdiff.GetYaxis().SetTitleSize(0.02)
zVertexdiff.SetLineColor(0)
zVertexdiff.SetMarkerColor(0)
zVertexdiff.SetMarkerStyle(20)

c = TCanvas("c", "c", 1000, 1000)
zVertexdiff.Draw()
c.Draw()
c.SaveAs("z_vertex_diffs.png")
```
