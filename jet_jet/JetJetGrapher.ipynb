# Histogram addition


```python
# Histogram-adding function
# Precondition: added_to and added_from are 1-D histograms that have the same dimensions
def Addto1D(added_to, added_from):
    x_bin_min = added_to.GetXaxis().FindBin(added_to.GetXaxis().GetXmin())
    x_bin_max = added_to.GetXaxis().FindBin(added_to.GetXaxis().GetXmax())
    i = x_bin_min
    while(i <= x_bin_max):
        added_to.SetBinContent(i, added_to.GetBinContent(i)+added_from.GetBinContent(i))
        
        # Error propogation
        added_to.SetBinError(i, TMath.Sqrt((added_to.GetBinError(i)*added_to.GetBinError(i)) + (added_from.GetBinError(i)*added_from.GetBinError(i))))
        
        i += 1
        
    return
```

```python
# Add all the MC histograms into one for each variable
Xj_MC = Xj_MC_pthat1.Clone()
Addto1D(Xj_MC, Xj_MC_pthat2)
Addto1D(Xj_MC, Xj_MC_pthat3)
Addto1D(Xj_MC, Xj_MC_pthat4)
Xjtruth_MC = Xjtruth_MC_pthat1.Clone()
Addto1D(Xjtruth_MC, Xjtruth_MC_pthat2)
Addto1D(Xjtruth_MC, Xjtruth_MC_pthat3)
Addto1D(Xjtruth_MC, Xjtruth_MC_pthat4)

dPhi_MC = dPhi_MC_pthat1.Clone()
Addto1D(dPhi_MC, dPhi_MC_pthat2)
Addto1D(dPhi_MC, dPhi_MC_pthat3)
Addto1D(dPhi_MC, dPhi_MC_pthat4)
dPhitruth_MC = dPhitruth_MC_pthat1.Clone()
Addto1D(dPhitruth_MC, dPhitruth_MC_pthat2)
Addto1D(dPhitruth_MC, dPhitruth_MC_pthat3)
Addto1D(dPhitruth_MC, dPhitruth_MC_pthat4)

ptD_MC = ptD_MC_pthat1.Clone()
Addto1D(ptD_MC, ptD_MC_pthat2)
Addto1D(ptD_MC, ptD_MC_pthat3)
Addto1D(ptD_MC, ptD_MC_pthat4)

Multiplicity_MC = Multiplicity_MC_pthat1.Clone()
Addto1D(Multiplicity_MC, Multiplicity_MC_pthat2)
Addto1D(Multiplicity_MC, Multiplicity_MC_pthat3)
Addto1D(Multiplicity_MC, Multiplicity_MC_pthat4)

dEta_MC = dEta_MC_pthat1.Clone()
Addto1D(dEta_MC, dEta_MC_pthat2)
Addto1D(dEta_MC, dEta_MC_pthat3)
Addto1D(dEta_MC, dEta_MC_pthat4)
dEtatruth_MC = dEtatruth_MC_pthat1.Clone()
Addto1D(dEtatruth_MC, dEtatruth_MC_pthat2)
Addto1D(dEtatruth_MC, dEtatruth_MC_pthat3)
Addto1D(dEtatruth_MC, dEtatruth_MC_pthat4)

AvgEta_MC = AvgEta_MC_pthat1.Clone()
Addto1D(AvgEta_MC, AvgEta_MC_pthat2)
Addto1D(AvgEta_MC, AvgEta_MC_pthat3)
Addto1D(AvgEta_MC, AvgEta_MC_pthat4)
AvgEtatruth_MC = AvgEtatruth_MC_pthat1.Clone()
Addto1D(AvgEtatruth_MC, AvgEtatruth_MC_pthat2)
Addto1D(AvgEtatruth_MC, AvgEtatruth_MC_pthat3)
Addto1D(AvgEtatruth_MC, AvgEtatruth_MC_pthat4)

leading_pT_MC = leading_pT_MC_pthat1.Clone()
Addto1D(leading_pT_MC, leading_pT_MC_pthat2)
Addto1D(leading_pT_MC, leading_pT_MC_pthat3)
Addto1D(leading_pT_MC, leading_pT_MC_pthat4)
leading_PTtruth_MC = leading_PTtruth_MC_pthat1.Clone()
Addto1D(leading_PTtruth_MC, leading_PTtruth_MC_pthat2)
Addto1D(leading_PTtruth_MC, leading_PTtruth_MC_pthat3)
Addto1D(leading_PTtruth_MC, leading_PTtruth_MC_pthat4)

leading_Eta_MC = leading_Eta_MC_pthat1.Clone()
Addto1D(leading_Eta_MC, leading_Eta_MC_pthat2)
Addto1D(leading_Eta_MC, leading_Eta_MC_pthat3)
Addto1D(leading_Eta_MC, leading_Eta_MC_pthat4)
leading_Etatruth_MC = leading_Etatruth_MC_pthat1.Clone()
Addto1D(leading_Etatruth_MC, leading_Etatruth_MC_pthat2)
Addto1D(leading_Etatruth_MC, leading_Etatruth_MC_pthat3)
Addto1D(leading_Etatruth_MC, leading_Etatruth_MC_pthat4)

leading_Phi_MC = leading_Phi_MC_pthat1.Clone()
Addto1D(leading_Phi_MC, leading_Phi_MC_pthat2)
Addto1D(leading_Phi_MC, leading_Phi_MC_pthat3)
Addto1D(leading_Phi_MC, leading_Phi_MC_pthat4)
leading_Phitruth_MC = leading_Phitruth_MC_pthat1.Clone()
Addto1D(leading_Phitruth_MC, leading_Phitruth_MC_pthat2)
Addto1D(leading_Phitruth_MC, leading_Phitruth_MC_pthat3)
Addto1D(leading_Phitruth_MC, leading_Phitruth_MC_pthat4)
```
