# This notebook creates graphs of various jet-jet PDFs from the output of jet_jet.cc

```python
import ROOT
from ROOT import TH1, TCanvas, TFile, TLatex, TLegend
```

```python
# Load the files
MinBias13cData = TFile("JetJet_config_13c.root", "READ")
MCData = TFile("JetJet_config_17g6a3_pthat1.root", "READ")

# Load the data
Xj_13c = MinBias13cData.Get("hjet_Xj")
Xj_MC = MCData.Get("hjet_Xj")
Xjtruth_MC = MCData.Get("hjet_Xj_truth")

dPhi_13c = MinBias13cData.Get("hjet_dPhi")
dPhi_MC = MCData.Get("hjet_dPhi")
dPhitruth_MC = MCData.Get("hjet_dPhi_truth")

ptD_13c = MinBias13cData.Get("hjet_ptD")
ptD_MC = MCData.Get("hjet_ptD")

Multiplicity_13c = MinBias13cData.Get("hjet_Multiplicity")
Multiplicity_MC = MCData.Get("hjet_Multiplicity")

dEta_13c = MinBias13cData.Get("hjet_dEta")
dEta_MC = MCData.Get("hjet_dEta")
dEtatruth_MC = MCData.Get("hjet_dEta_truth")

AvgEta_13c = MinBias13cData.Get("hjet_AvgEta;1")
AvgEta_MC = MCData.Get("hjet_AvgEta")
AvgEtatruth_MC = MCData.Get("hjet_AvgEta_truth")
```

```python
# Graph Xj
Xj_13c.SetTitle("Xj distribution; Xj; Portion of total sample")
Xj_13c.SetMaximum(0.05)
Xj_13c.GetYaxis().SetTitleOffset(1)
Xj_13c.GetYaxis().SetLabelSize(0.02)
Xj_13c.GetYaxis().SetTitleSize(0.04)
Xj_13c.SetLineColor(1)
Xj_13c.SetMarkerColor(1)
Xj_13c.SetMarkerStyle(20)
Xj_MC.SetLineColor(2)
Xj_MC.SetMarkerColor(2)
Xj_MC.SetMarkerStyle(20)
Xjtruth_MC.SetLineColor(3)
Xjtruth_MC.SetMarkerColor(3)
Xjtruth_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
Xj_13c.Draw()
Xj_MC.Draw("same")
Xjtruth_MC.Draw("HIST same")
c.Draw()

legend = TLegend(0.78, 0.77, 0.98, .94)
legend.AddEntry(Xj_13c, "13c Min-bias Data", "l")
legend.AddEntry(Xj_MC, "MC results", "l")
legend.AddEntry(Xjtruth_MC, "MC truth", "l")
legend.Draw()
c.SaveAs("Xj.png")
```

```python
# Graph dPhi
dPhi_13c.SetTitle("#Delta #phi distribution; #Delta #phi (rads); Portion of total sample")
dPhi_13c.SetMaximum(0.035)
dPhi_13c.SetMinimum(0)
dPhi_13c.GetYaxis().SetTitleOffset(1)
dPhi_13c.GetYaxis().SetLabelSize(0.02)
dPhi_13c.GetYaxis().SetTitleSize(0.04)
dPhi_13c.SetLineColor(1)
dPhi_13c.SetMarkerColor(1)
dPhi_13c.SetMarkerStyle(20)
dPhi_MC.SetLineColor(2)
dPhi_MC.SetMarkerColor(2)
dPhi_MC.SetMarkerStyle(20)
dPhitruth_MC.SetLineColor(3)
dPhitruth_MC.SetMarkerColor(3)
dPhitruth_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
dPhi_13c.Draw()
dPhi_MC.Draw("same")
dPhitruth_MC.Draw("HIST same")
c.Draw()

legend.Draw()
c.SaveAs("dPhi.png")
```

```python
# Graph delta eta
dEta_13c.SetTitle("#Delta #eta distribution; #Delta #eta; Portion of total sample")
dEta_13c.SetMaximum(0.06)
dEta_13c.SetMinimum(0)
dEta_13c.GetYaxis().SetTitleOffset(1)
dEta_13c.GetYaxis().SetLabelSize(0.02)
dEta_13c.GetYaxis().SetTitleSize(0.04)
dEta_13c.GetXaxis().SetRangeUser(0, 1.2)
dEta_13c.SetLineColor(1)
dEta_13c.SetMarkerColor(1)
dEta_13c.SetMarkerStyle(20)
dEta_MC.SetLineColor(2)
dEta_MC.SetMarkerColor(2)
dEta_MC.SetMarkerStyle(20)
dEtatruth_MC.SetLineColor(3)
dEtatruth_MC.SetMarkerColor(3)
dEtatruth_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
dEta_13c.Draw()
dEta_MC.Draw("same")
dEtatruth_MC.Draw("HIST same")
c.Draw()

legend.Draw()
c.SaveAs("dEta.png")
```

```python
# Graph Average Eta
AvgEta_13c.SetTitle("Average #eta distribution; #eta_{Avg}; Portion of total sample")
AvgEta_13c.SetMinimum(0)
AvgEta_13c.SetMaximum(0.03)
AvgEta_13c.GetYaxis().SetTitleOffset(1)
AvgEta_13c.GetYaxis().SetLabelSize(0.02)
AvgEta_13c.GetYaxis().SetTitleSize(0.04)
AvgEta_13c.SetLineColor(1)
AvgEta_13c.SetMarkerColor(1)
AvgEta_13c.SetMarkerStyle(20)
AvgEta_MC.SetLineColor(2)
AvgEta_MC.SetMarkerColor(2)
AvgEta_MC.SetMarkerStyle(20)
AvgEtatruth_MC.SetLineColor(3)
AvgEtatruth_MC.SetMarkerColor(3)
AvgEtatruth_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
AvgEta_13c.Draw()
AvgEta_MC.Draw("same")
AvgEtatruth_MC.Draw("HIST same")
c.Draw()

legend.Draw()
c.SaveAs("AvgEta.png")
```

```python
# Graph ptD
ptD_13c.SetTitle("p_{T}D distribution; p_{T}D (GeV); Portion of total sample")
ptD_13c.SetMaximum(0.015)
ptD_13c.SetMinimum(0)
ptD_13c.GetYaxis().SetTitleOffset(1)
ptD_13c.GetYaxis().SetLabelSize(0.02)
ptD_13c.GetYaxis().SetTitleSize(0.04)
ptD_13c.SetLineColor(1)
ptD_13c.SetMarkerColor(1)
ptD_13c.SetMarkerStyle(20)
ptD_MC.SetLineColor(2)
ptD_MC.SetMarkerColor(2)
ptD_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
ptD_13c.Draw()
ptD_MC.Draw("same")
c.Draw()

legend1 = TLegend(0.78, 0.77, 0.98, .94)
legend1.AddEntry(ptD_13c, "13c Min-bias Data", "l")
legend1.AddEntry(ptD_MC, "MC results", "l")
legend1.Draw()
c.SaveAs("ptD.png")
```

```python
# Graph Multiplicity
Multiplicity_13c.SetTitle("Multiplicity distribution; Multiplicity; Portion of total sample")
Multiplicity_13c.GetYaxis().SetTitleOffset(1)
Multiplicity_13c.GetYaxis().SetLabelSize(0.02)
Multiplicity_13c.GetYaxis().SetTitleSize(0.04)
Multiplicity_13c.SetLineColor(1)
Multiplicity_13c.SetMarkerColor(1)
Multiplicity_13c.SetMarkerStyle(20)
Multiplicity_MC.SetLineColor(2)
Multiplicity_MC.SetMarkerColor(2)
Multiplicity_MC.SetMarkerStyle(20)

c = TCanvas("c", "c", 1600, 1600)
Multiplicity_13c.Draw()
Multiplicity_MC.Draw("same")
c.Draw()

legend1.Draw()
c.SaveAs("Mult.png")
```
